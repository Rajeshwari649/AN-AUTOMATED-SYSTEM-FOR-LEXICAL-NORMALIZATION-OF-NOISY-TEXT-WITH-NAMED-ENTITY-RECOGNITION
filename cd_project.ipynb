```python
import string
import spacy
from __future__ import with_statement

nlp = spacy.load('en_core_web_sm')
# load text
filename = input("Enter your name of the dataset: ")
with open (filename + ".txt","r") as file:
    text = file.read()
    sentence= text.split('.')

print("THE PARAGRAPH SPLIT INTO SENTENCES")
print(sentence)

table = str.maketrans('', '', string.punctuation)
stripped = [s.translate(table) for s in sentence]
print("THE PARAGRAPH AFTER REMOVING PUNCTUATION")
print(stripped)

print("THE PARAGRAPH AFTER TOKENIZATION")
token_list = []
for token in stripped:
        token_list.append(token.split())

print(token_list)
with open('output.txt', 'w') as f:
    for _list in token_list:
         for _string in _list:
            f.write(str(_string) + " ")

```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
keywords = set()
with open('file1.txt') as list_file:
    for line in list_file:
        if line.strip():
            keywords.add(line.strip())

with open('abbr.txt') as master_file:
    with open('nn.txt', 'w') as search_results:
        for line in master_file:
            if set(line.split()[:-1]) & keywords:
                search_results.write(line)
f1 = open("nn.txt", "r") 
f2 = open("file1.txt", "r") 

file1_raw = f1.read()
file2_raw = f2.read()

file1_words = file1_raw.split()
file2_words = file2_raw.split()

result = set(file1_words).difference(set(file2_words))

with open('listfile.txt', 'w') as filehandle:
    for listitem in result:
        filehandle.write('%s\n' % listitem)

assembled = " ".join(result)
doc = nlp(assembled)

nouns = []
for token in doc:
   if token.pos_ == "NOUN":
        nouns.append(token)

postag = open("nouns.txt", "a")
for word in nouns:
    postag.write(str(word) + "\n")
postag.close()

# lines = open('nouns.txt', 'r').readlines()
# lines_set = set(lines)
# out  = open('nouns.txt', 'w')
# for line in lines_set:
#     out.write(line)
```
